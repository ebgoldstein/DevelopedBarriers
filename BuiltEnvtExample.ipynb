```python
#Import the packages we need
import osmnx as ox
import geopandas
```

```python
#read in an example shapefile. this one is from Bogue Banks NC
Island = geopandas.read_file("compiled_boundary_poly_shp/NC4_geo.shp")
```

```python
#look at the data just make sure it is what i expect
Island.head
```

```python
#Plot up the outline
Island.plot()
```

```python
#extract just the geometry (shapely object) part
isle = Island['geometry'].iloc[0]
#Check to see if the polygon is valid (w/o holes, etc.)
isle.is_valid
```

```python
#since it is not, we clean it with a buffer
clean = isle.buffer(0)
clean.is_valid
```

```python
#plot it up
clean
```

```python
#pull the roads with OSMnx and plot it up
RoadNet = ox.graph_from_polygon(clean)
ox.plot_graph(RoadNet)
```
